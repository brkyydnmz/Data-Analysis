```python
#############################################
# Toplulaştırma ve Gruplama (Aggregation & Grouping)
#############################################

# - count()
# - first()
# - last()
# - mean()
# - median()
# - min()
# - max()
# - std()
# - var()
# - sum()
# - pivot table

import pandas as pd
import seaborn as sns
pd.set_option('display.max_columns', None)
df = sns.load_dataset("titanic")
df.head()

```

```python
df["age"].mean()

```

```python
df.groupby("sex")["age"].mean()

```

```python
df.groupby("sex").agg({"age": "mean"})
df.groupby("sex").agg({"age": ["mean", "sum"]})

```

```python
df.groupby("sex").agg({"age": ["mean", "sum"],"survived": "mean"})


```

```python
df.groupby(["sex", "embark_town"]).agg({"age": ["mean"],"survived": "mean"})

```

```python
df.groupby(["sex", "embark_town", "class"]).agg({"age": ["mean"],"survived": "mean"})


```

```python
df.groupby(["sex", "embark_town", "class"]).agg({
    "age": ["mean"],
    "survived": "mean",
    "sex": "count"})
```

```python
#############################################
# Apply ve Lambda
#############################################
import pandas as pd
import seaborn as sns
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
df = sns.load_dataset("titanic")
df.head()

```

```python
(df["age"]/10).head()

```

```python
df["age2"] = df["age"]*2
(df["age2"]/10).head()

```

```python
df["age3"] = df["age"]*5
(df["age3"]/10).head()
```

```python
for col in df.columns:
    if "age" in col:
        print(col)
```

```python
for col in df.columns:
    if "age" in col:
        print((df[col]/10).head())

```

```python
for col in df.columns:
    if "age" in col:
        df[col] = df[col]/10

df.head()

```

```python
df[["age", "age2", "age3"]].apply(lambda x: x/10).head()

```

```python
df.loc[:, df.columns.str.contains("age")].apply(lambda x: x/10).head()

```

```python
df.loc[:, df.columns.str.contains("age")].apply(lambda x: (x - x.mean()) / x.std()).head()

```

```python
def standart_scaler(col_name):
    return (col_name - col_name.mean()) / col_name.std()

```

```python
df.loc[:, df.columns.str.contains("age")].apply(standart_scaler).head()

```

```python
# df.loc[:, ["age","age2","age3"]] = df.loc[:, df.columns.str.contains("age")].apply(standart_scaler)

df.loc[:, df.columns.str.contains("age")] = df.loc[:, df.columns.str.contains("age")].apply(standart_scaler)

df.head()
```

```python
#############################################
# Birleştirme (Join) İşlemleri
#############################################
import numpy as np
import pandas as pd
m = np.random.randint(1, 30, size=(5, 3))
m
```

```python
df1 = pd.DataFrame(m, columns=["var1", "var2", "var3"])
df2 = df1 + 99

pd.concat([df1, df2])

```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
#######################
# Merge ile Birleştirme İşlemleri
#######################

df1 = pd.DataFrame({'employees': ['john', 'dennis', 'mark', 'maria'],
                    'group': ['accounting', 'engineering', 'engineering', 'hr']})

df2 = pd.DataFrame({'employees': ['mark', 'john', 'dennis', 'maria'],
                    'start_date': [2010, 2009, 2014, 2019]})

pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on="employees")

```

```python
# Amaç: Her çalışanın müdürünün bilgisine erişmek istiyoruz.
df3 = pd.merge(df1, df2)

df4 = pd.DataFrame({'group': ['accounting', 'engineering', 'hr'],
                    'manager': ['Caner', 'Mustafa', 'Berkcan']})

pd.merge(df3, df4)
```
